```python
import cv2
import numpy as np

# Load the image
path="C:\\Users\\yedla\\Dropbox\\PC\\Desktop\\Fabric_defect_detection\\Task4 data\\Fabric2.jpg"
image = cv2.imread(path)

# Convert color image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Denoise the grayscale image using GaussianBlur or bilateralFilter
# denoised_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
# denoised_image=cv2.bitwise_not(denoised_image)
# Alternatively, use bilateralFilter for noise removal while preserving edges
denoised_image_bilateral = cv2.bilateralFilter(gray_image, 9, 75, 75)
denoised_image_bilateral=cv2.bitwise_not(denoised_image_bilateral)

sharpened_image = cv2.addWeighted(gray_image, 1.5, denoised_image_bilateral, -0.5, 0)
sharpened_image=cv2.bitwise_not(sharpened_image)
# Apply histogram equalization for enhancing the defects
# clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

# # Apply CLAHE to enhance local contrast
# enhanced_image = clahe.apply(sharpened_image)
#enhanced_image = cv2.equalizeHist(denoised_image_bilateral)

# Display images (optional)
# cv2.imshow('Original Image', gray_image)

# cv2.imshow('Bilateral Image', denoised_image_bilateral)
# cv2.imshow('Sharpen', sharpened_image)
# cv2.imshow('Enhance', enhanced_image)
kernel = np.ones((3, 3), np.uint8)
seed_point = (256, 256)
new_value = 0
closed_image = cv2.morphologyEx(sharpened_image, cv2.MORPH_CLOSE, kernel)
closed_image=cv2.bitwise_not(closed_image)
# _, final_segmented_image = cv2.threshold(closed_image, 0, 255, cv2.THRESH_BINARY)
cv2.floodFill(closed_image, None, seedPoint=seed_point, newVal=new_value, loDiff=(5, 5, 5, 5), upDiff=(5, 5, 5, 5))
cv2.imshow('Flood-Filled Image', closed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
